<h3>We start by downloading the data and load it using the pandas library</h3>
<br>


```python
import pandas as pd
train = pd.read_csv("C:\\Users\\houss\\Documents\\Drive\\kaggle\\HousePricing\\train.csv") # change your own path 
test = pd.read_csv("C:\\Users\\houss\\Documents\\Drive\\kaggle\\HousePricing\\test.csv")
```

<h3> Please note that the test dataset doesn't have a <mark>'SalePrice'</mark> column. Actually this is the goal of the competiton : you'll predict the values of this column. So the model training and its evaluation will be on the train dataset (use the train_test_split method on the train dataset)

For this exemple we'll focus only on the 'GrLivArea' feature

```python
X = train[['GrLivArea']]
y= train ['SalePrice']
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #initialise the model
regressor.fit(X,y) #fit the model
```

<h3>Now it's time to predict the <mark>'SalePrice'</mark> values using the test dataset 

```python
y_pred = regressor.predict(test[['GrLivArea']])
```

<h3> if we look to the submission sample required, we see that we need to submit a .csv file that contain the Id column and the predicted sale price </h3>
    


```python
test['SalePrice'] = y_pred
submission = test[['Id', 'SalePrice']]
submission.to_csv('sub.csv', index=False)
```
